```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from keras.models import Sequential, Model, load_model
import os
from keras.preprocessing import image

from keras.applications.vgg19 import (
    VGG19, preprocess_input, decode_predictions)
import numpy as np
```

```python
labels = pd.read_csv("labels.csv")
```

```python
names = ['Avocado', 'Broccoli', 'Brussels Sprouts', 'Cabbage', 'Carrot',
       'Celery', 'Coliflower', 'Corn', 'Eggplant', 'Green Beans',
       'Green Tomato', 'Mushroom', 'Okra', 'Onion', 'Potato', 'Pumpkin',
       'Radish', 'Red Tomato', 'Spinach', 'Spring Onion', 'beet', 'cucumber',
       'peas', 'sweetpotato']
```

```python
model = load_model("vgg19_pictoreceipe_1.h5")
```

```python
images = ["cucumber_1.jpg", "beet_1.jpg", "celery5.jpg", "okra_1.jpg", "okra_2.jpg", "testb.jpg", "testc.jpg", "testo.jpg", "pumpkin_1.jpg", "Cabbage1.jpg", "sweetpotato_1.jpg", "cabbage_2.jpg", "cabbage_3.jpg"]
```

```python
image_path_array = []
for img in images:
    image_path = os.path.join("DataSet/single_prediction/" + img)
    image_path_array.append(image_path)
print(image_path_array)
```

```python
image_path = image_path_array[0]
img = image.load_img(image_path, target_size=(224, 224))
plt.imshow(img)
```

```python
test_image = image.img_to_array(img)
test_image = np.expand_dims(test_image, axis=0)
x = preprocess_input(test_image)
```

```python
plt.imshow(x[0])
predictions = model.predict(x)
names[predictions.argmax(axis=1)[0]]
```
